```python
df = df.rename(columns={'PM2.5': 'pm25'})
```

```python
y = pd.DataFrame(df, columns= ['pm25'])

print (y)

X= df.loc[ : , df.columns != 'pm25']
print(X)
```

# **MLP**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(
    X,y,test_size=0.33, random_state=50)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Train one model with raw data to establish a reference.
# In this case, we will train a MLP
from sklearn.neural_network import MLPRegressor

#Initializing the MLPRegressor
regr = MLPRegressor(hidden_layer_sizes=(100,50,10), max_iter=10,
                           activation = 'relu',solver='adam',random_state=1)

#Fitting the training data to the network
regr.fit(X_train, y_train.values.ravel())

#Predicting y for X_val
y_pred = regr.predict(X_test)

```

```python
from sklearn import metrics
```

```python
from sklearn import metrics

#calculte r2 
print('R2 =' ,metrics.r2_score(y_test, y_pred))
```

```python
from sklearn.metrics import mean_absolute_error as mae

#calculate MAE
mae = mae(y_test, y_pred)
print('MAE = ', mae)
```

```python
prediction = [4, 4, 7, 300, 77, -0.3, 1023, -18.8, 0, 4.4]
yhat = regr.predict([prediction])
# summarize the prediction
print('Predicted: %s' % yhat[0])
```
